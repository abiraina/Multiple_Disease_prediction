```python
# Check for missing values in each column
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Encode categorical variables
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for col in data.select_dtypes(include="object").columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
X = data.drop('status', axis=1)
y = data['status']
model = LogisticRegression(max_iter=1000)
rfe = RFE(model, n_features_to_select=10)
fit = rfe.fit(X, y)

selected_features = X.columns[fit.support_]
print("Selected Features using RFE:", selected_features.tolist())
```

```python
# Features and target
X = data[['MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'MDVP:APQ', 'Shimmer:DDA', 'RPDE', 'DFA', 'spread1', 'spread2', 'D2', 'PPE']]
y = data["status"]
```

```python
# Scale features 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
y_train.value_counts()
```

```python
# Apply SMOTE to balance the training data
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(X_train_bal,y_train_bal)
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train_bal,y_train_bal)
y_pred=dt.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10)#parameter how many decision tree has to be generator
rf.fit(X_train_bal,y_train_bal)
y_pred=rf.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Plot ROC Curve
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
y_proba = rf.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_proba)
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
plt.figure()
plt.plot(fpr, tpr, label=f"RandomForest (AUC = {roc_auc_score(y_test, y_proba):.2f})", color='purple')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(" RandomForest- ROC Curve")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
# Plot Confusion Matrix
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap="Purples")
plt.title("RandomForest - Confusion Matrix")
plt.show()
```
